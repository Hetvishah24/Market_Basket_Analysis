```python
from apyori import apriori
```

```python
#loading market basket dataset..

df = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
```

```python
df.fillna(0,inplace=True)
```

```python
items = []

for i in range(0,len(df)):
    items.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
items[0]
```

```python
#applying apriori algorithm
rules = apriori(items, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
rules = list(rules)
```

```python
rules
```

```python
result = pd.DataFrame(rules)
```

```python
result
```

```python
item1 = []
item2 = []
confi = []
lift = []
support = result.support

for i in range(result.shape[0]):
    statistics = result['ordered_statistics'][i][0]
    item1.append(list(statistics[0]))
    item2.append(list(statistics[1]))
    confi.append(statistics[2])
    lift.append(statistics[3])
```

```python
dict = {'Item1': item1, 'Item2': item2, 'Support': support, 'Confidence':confi, 'Lift':lift} 
    
df = pd.DataFrame(dict)
```
